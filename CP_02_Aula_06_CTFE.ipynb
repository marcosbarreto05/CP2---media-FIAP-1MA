```python
# primeiro semestre
# calculo de media dos Checkpoints (CP1, CP2, CP3)
# 3 CPs obrigatorios ---> input
# eliminar a menor nota
# calcular media das duas maiores notas (MCP_1SEM)
```

```python
CP1_S1, CP2_S1, CP3_S1 = input('Insira as notas dos checkpoints: ').split()
CP1_S1 = float(CP1_S1)
CP2_S1 = float(CP2_S1)
CP3_S1 = float(CP3_S1)

if(CP1_S1 and CP2_S1 > CP3_S1):
  MCP_1SEM = (CP1_S1 + CP2_S1) / 2
  print(f'Media: {MCP_1SEM:.1f}')
elif(CP1_S1 and CP3_S1 > CP2_S1):
   MCP_1SEM = (CP1_S1 + CP3_S1) / 2
   print(f'Media: {MCP_1SEM:.1f}')
else:
   MCP_1SEM = (CP2_S1 + CP3_S1) / 2
   print(f'Media: {MCP_1SEM:.1f}')
```
