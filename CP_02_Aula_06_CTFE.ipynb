Desafio: Cálculo da média anual FIAP

```python
# primeiro semestre
# calculo de media dos Checkpoints (CP1, CP2, CP3)
# 3 CPs obrigatorios ---> input
# eliminar a menor nota
# calcular media das duas maiores notas (MCP_1SEM)
```

```python
CP1_1SEM, CP2_1SEM, CP3_1SEM = input('Insira as notas dos checkpoints: ').split()
CP1_1SEM = float(CP1_1SEM)
CP2_1SEM = float(CP2_1SEM)
CP3_1SEM = float(CP3_1SEM)

if(CP1_1SEM and CP2_1SEM > CP3_1SEM):
  MCP_1SEM = (CP1_1SEM + CP2_1SEM) / 2
  print(f'Media: {MCP_1SEM:.1f}')
elif(CP1_1SEM and CP3_1SEM > CP2_1SEM):
   MCP_1SEM = (CP1_1SEM + CP3_1SEM) / 2
   print(f'Media: {MCP_1SEM:.1f}')
else:
   MCP_1SEM = (CP2_1SEM + CP3_1SEM) / 2
   print(f'Media: {MCP_1SEM:.1f}')
```

```python
# Calculo de notas de SPRINTs (SP1 e SP2)
# Calcular media das duas notas (SP_1SEM)
```

```python
SP1_1SEM, SP2_1SEM = input('Insira as notas dos sprints: ').split()
SP1_1SEM = float(SP1_1SEM)
SP2_1SEM = float(SP2_1SEM)

SP_1SEM = (SP1_1SEM + SP2_1SEM) / 2
print(f'Media: {SP_1SEM:.1f}')
```

```python
# Mostrar quanto falta para o aluno passar direto
# Calculo de notas de Global Solution (GS)
# Inserir a nota.
# Caso não haja nota de GS, o aluno deverá inserir sua nota de SUB
```

```python
RES_1 = 14 - (0.2*SP_1SEM + 0.2*MCP_1SEM)
RES_1 = RES_1*0.6 - 4
print(f'Falta: {RES_1:.1f}')

GS1 = input('Insira a nota da Global Solution: ')
GS1 = float(GS1)

if(GS1 == 0):
  SUB1 = input('Insira a nota da Substitutiva: ')
  SUB1 = float(SUB1)
```

```python
# Calculo da media semestral 1 (MS1)
if(GS1 == 0):
  GS1 = SUB1
  MS1 = 0.6*GS1 + 0.2*SP_1SEM + 0.2*MCP_1SEM
  print(f'Media: {MS1:.1f}')
else:
  MS1 = 0.6*GS1 + 0.2*SP_1SEM + 0.2*MCP_1SEM
  print(f'Media: {MS1:.1f}')
```

```python
# Repetir para o segundo semestre
```

```python
# segundo semestre
# calculo de media dos Checkpoints (CP1, CP2, CP3)

CP1_2SEM, CP2_2SEM, CP3_2SEM = input('Insira as notas dos checkpoints: ').split()
CP1_2SEM = float(CP1_2SEM)
CP2_2SEM = float(CP2_2SEM)
CP3_2SEM = float(CP3_2SEM)

if(CP1_2SEM and CP2_2SEM > CP3_2SEM):
  MCP_2SEM = (CP1_2SEM + CP2_2SEM) / 2
  print(f'Media: {MCP_2SEM:.1f}')
elif(CP1_2SEM and CP3_2SEM > CP2_2SEM):
   MCP_2SEM = (CP1_2SEM + CP3_2SEM) / 2
   print(f'Media: {MCP_2SEM:.1f}')
else:
   MCP_2SEM = (CP2_2SEM + CP3_2SEM) / 2
   print(f'Media: {MCP_2SEM:.1f}')

# Calculo de notas de SPRINTs (SP1 e SP2)

SP1_2SEM, SP2_2SEM = input('Insira as notas dos sprints: ').split()
SP1_2SEM = float(SP1_2SEM)
SP2_2SEM = float(SP2_2SEM)

SP_2SEM = (SP1_2SEM + SP2_2SEM) / 2
print(f'Media: {SP_2SEM:.1f}')

# Calculo de notas de Global Solution (GS)

RES_2 = 14 - (0.2*SP_2SEM + 0.2*MCP_2SEM)
RES_2 = RES_2*0.6 - 4
print(f'Falta: {RES_2:.1f}')

GS2 = input('Insira a nota da Global Solution: ')
GS2 = float(GS2)

if(GS2 == 0):
  SUB2 = input('Insira a nota da Substitutiva: ')
  SUB2 = float(SUB2)

# Calculo da media semestral 1 (MS1)

if(GS2 == 0):
  GS2 = SUB2
  MS2 = 0.6*GS2 + 0.2*SP_2SEM + 0.2*MCP_2SEM
  print(f'Media: {MS2:.1f}')
else:
  MS2 = 0.6*GS2 + 0.2*SP_2SEM + 0.2*MCP_2SEM
  print(f'Media: {MS2:.1f}')
```

```python
# Media Final FIAP
MF = 0.4*MS1 + 0.6*MS2
print(f'Media Final: {MF:.1f}')
```

```python
# Apresentar se o aluno esta aprovado, reprovado ou de exame.
# Aprovado >= 6.0
# Exame >= 4.0 e < 6.0
# Reprovado < 4.0
# Se estiver de exame, inserir nota do exame
# Calcula MF2 = (EX + MF) / 2

if(MF == 4.0 or MF <= 5.9):
  print('Aluno em exame')
  EX = input('Insira a nota do Exame: ')
  EX = float(EX)
  MF2 = (EX + MF) / 2
  print(f'Media Final: {MF2:.1f}')
  if(MF2 >= 6):
    print('Aluno aprovado')
  else:
    print('Aluno reprovado')
elif(MF >= 6.0):
  print('Aluno aprovado')
else:
  print('Aluno reprovado')
```
